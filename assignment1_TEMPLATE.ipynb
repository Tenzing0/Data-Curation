```python
#Task 1: Data Preparation
# "You will start by loading the CSV data from the file (using appropriate pandas functions) and checking whether the loaded data is equivalent to the data in the source CSV file.

#Importing all the import packages required 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
Dataset="Automobile.csv"
#Loading the dataset and giving column names
Automobile = pd.read_csv(Dataset,sep='#',decimal='.', header=None,
                         names=['Symboling',
'Normalized-losses',
'Make',
'Fuel-type',
'Aspiration',
'Num-of-doors',
'Body-style',
'Drive-wheels',
'Engine-location',
'Wheel-base',
'Length',
'Width',
'Height',
'Curb-weight',
'Engine-type',
'Num-of-cylinders',
'Engine-size',
'Fuel-system',
'Bore',
'Stroke',
'Compression-ratio',
'Horsepower',
'Peak-rpm',
'City-mpg',
'Highway-mpg',
'Price'])

# Then, you need to clean the data by using the knowledge we taught in the lectures. You need to deal with all the potential issues/errors in the data appropriately (such as: typos, extra whitespaces, sanity checks for impossible values, and missing values etc). "

# Please structure code as follows: 
# always provide one line of comments to explain the purpose of the code, e.g. load the data, checking the equivalent to original data, checking typos (do this for each other types of errors)
```

```python
#Removing all duplicate observations 
Automobile = Automobile.drop_duplicates()
```

```python
#Checking the various data types present in the dataset
Automobile.dtypes
```

```python
# There seems to be values going beyond the expert's threshold for Symboling,Normalized-Losses,Drive-Wheels and Price.
Automobile.describe()
```

```python
#Replacing typos and redundant white Space

Automobile['Make'].replace('vol00112ov','volvo',inplace=True)
Automobile['Make'] = Automobile['Make'].str.strip() 
Automobile['Fuel-type'].replace('Gas','gas',inplace=True)
Automobile['Fuel-type']=Automobile['Fuel-type'].str.strip()
Automobile['Fuel-type'].replace('Diesel','diesel',inplace=True)
Automobile['Aspiration'].replace('turrrrbo','turbo',inplace=True)
Automobile['Aspiration'].replace('Std','std',inplace=True)
Automobile['Aspiration']=Automobile['Aspiration'].str.strip()
Automobile['Body-style']=Automobile['Body-style'].str.strip()
Automobile['Body-style'].replace('Sedan','sedan',inplace=True)
Automobile['Body-style'].replace('Wagon','wagon',inplace=True)
Automobile['Drive-wheels'] = Automobile['Drive-wheels'].str.strip()
Automobile['Drive-wheels'].replace('4wd','fwd',inplace=True)
Automobile['Drive-wheels'].replace('Fwd','fwd',inplace=True)
Automobile['Engine-location']=Automobile['Engine-location'].str.strip()
Automobile['Engine-location'].replace('Front','front',inplace=True)
Automobile['Engine-location'].replace('FRONT','front',inplace=True)
Automobile['Engine-location'].replace('REAR','rear',inplace=True)
Automobile['Engine-location'].replace('Rear','rear',inplace=True)
Automobile['Engine-type'].replace('Ohc','ohc',inplace=True)
Automobile['Num-of-cylinders'] = Automobile['Num-of-cylinders'].str.strip()
Automobile['Num-of-cylinders'].replace('Four','four',inplace=True)
Automobile['Fuel-system'].replace('Mpfi','mpfi',inplace=True)
Automobile['Fuel-system'] = Automobile['Fuel-system'].str.strip()
```

```python
# In the numerical data see below there seems to be values missing in Normalized-losses,Horsepower,Peak-rpm and price.
Automobile.isnull().sum()
```

```python

Automobile.loc[(Automobile['Normalized-losses'] <65)] #Checking for impossible values
Automobile['Normalized-losses'].replace(25, 65,inplace=True)#Replacing it with the nearest threshold

#Replacing the price of the vehicle with the standard price of the manufacturer
Automobile['Normalized-losses'] = Automobile['Normalized-losses'].fillna(Automobile.groupby('Make')['Normalized-losses'].transform('mean'))

#Using the for loop to replace  all the missing values of the columns by its mode
for column in ['Num-of-doors']:
    Automobile[column].fillna(Automobile[column].mode()[0], inplace=True)

    #Using the for loop to replace  all the missing values of the columns by its mean 
for column in ['Peak-rpm','Stroke','Bore','Horsepower','City-mpg','Normalized-losses']:
    Automobile[column].fillna(Automobile[column].mean(axis=0), inplace=True)
Automobile.isnull().sum()

#Replacing price of the Nan and zero values to the mean of the manufacturer(make)  of the Automobile.
Automobile['Price'] = Automobile['Price'].replace(0, np.nan)
Automobile['Price'] = Automobile['Price'].fillna(Automobile.groupby('Make')['Price'].transform('mean'))

#Capping it with the nearest value.
Automobile['Symboling'].replace(4, 3,inplace=True)
Automobile['Symboling'].value_counts()

#Changing the Symboling Column into an object as it is a ordinal variable
Automobile['Symboling']=Automobile['Symboling'].astype(object)

Automobile.isnull().sum()
```

```python
# Task 2: Data Exploration
# 1.  Nominal value - Body-Style
Automobile['Body-style'].value_counts().plot(kind='pie',autopct='%.2f')#Calculating the frequency and then plotting the pie chart
plt.title("Pie Chart representing Body-Style")#Putting a main label for the histogram
plt.show()#Displaying the plot
#2.Ordinal Value - Symboling  
Automobile['Symboling'].plot(kind='hist',bins=20)#Plotting the histogram
plt.title('Frequency Distribution of Symboling ')#Putting a main label for the histogram
plt.xlabel('Symboling')#Labelling the x axis
plt.show()# Displaying the plot
#3.Numerical Value - Horse Power
Automobile['Horsepower'].plot(kind='box')#Plotting the boxplot
plt.title('BoxPlot of the Horsepower')#Putting a main label for the boxplot
plt.show()# Displaying the plot
```

```python
# Task 2: Data Exploration
#Boxplot to Explore the relationship between Horsepower and body Style 
Automobile.dropna().boxplot(column='Price',by='Body-style')
plt.ylabel("Price")

#Boxplot to Explore the relationship between Price  and Symboling
Automobile.dropna().boxplot(column='Price',by='Symboling')
plt.ylabel("Price")
#Pivot Table to explore the relationship between Price v/s Symboling and BodyStyle 
pd.pivot_table(Automobile,index=['Body-style','Symboling'], values='Price').plot(kind='bar',color='blue')
plt.title("Price based on body style and Symboling")
plt.ylabel('Price')
plt.xlabel('Body style and Symboling');
```
